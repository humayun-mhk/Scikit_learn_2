```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
url = "https://raw.githubusercontent.com/selva86/datasets/master/Advertising.csv"
df = pd.read_csv(url)
print(df)
# Extract feature (TV ad budget) and target (Sales)
X = df[['TV']]  # Keep X as a 2D array for sklearn
Y = df['sales']

# Fit the model using scikit-learn
model = LinearRegression()
model.fit(X, Y)

# Predict sales using the model
Y_pred = model.predict(X)
results = pd.DataFrame({
    'TV Budget ($1000)': X['TV'],
    'Actual Sales': Y,
    'Predicted Sales': Y_pred
})

# Print the first few rows
print(results.head())
# Plot the results
plt.scatter(X, Y, color='blue', label="Actual Data")
plt.plot(X, Y_pred, color='red', label="Sklearn Regression Line")
plt.xlabel("TV Advertising Budget ($1000)")
plt.ylabel("Sales ($1000)")
plt.legend()
plt.title("Linear Regression using Sklearn")
plt.show()

# Print slope and intercept
print(f"Sklearn: Slope = {model.coef_[0]:.4f}, Intercept = {model.intercept_:.4f}")
```
